```python
import spacy
nlp = spacy.load("ja_ginza")
text = "隣の客はよく柿食う客だ。"
doc = nlp(text)
for token in doc:
  print("{}\t{}\t{}\t{}".format(token,token.lemma_,token.pos_,token.tag_))
```
